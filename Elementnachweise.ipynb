# Plattennachweise Test

```python
# Imports the "Platte" object from the "Elementnachweise.py" file into this notebook
from Elementnachweise import Platte
```

### Creating some ceiling-objects

```python
# Saves an "Platte" object with specific attributes into the variable "ceiling_1". "Richtung" describes the destination of the 4th reinforcement layer.
ceiling_1 = Platte(filename_innerforce='Schnittkraefte.csv', filename_raction='Flaechenreaktionen.csv', c_nom_mm=30, d_mm=300, Richtung='x',
                   d4_mm=12, a4_mm=150, d3_mm=12, a3_mm=150, s_mm=0, f_cd_Nmm2=16.5, tau_cd_Nmm2=1.00, Dmax_mm=32, f_sd_Nmm2=435, E_s_Nmm2=205000)
# Multiple variables with different attributes can be created
ceiling_2 = Platte(filename_innerforce='Schnittkraefte.csv', filename_raction='Flaechenreaktionen.csv', c_nom_mm=30, d_mm=300, Richtung='x',
                   d4_mm=14, a4_mm=150, d3_mm=14, a3_mm=150, s_mm=0, f_cd_Nmm2=16.5, tau_cd_Nmm2=1.00, Dmax_mm=32, f_sd_Nmm2=435, E_s_Nmm2=205000)
```

```python
# only last command-row is automatically printed. 
ceiling_1.m_3_Rd() # getting the bending resistance of the third layer
ceiling_1.m_4_Rd() # getting the bending resistance of the fouth layer
ceiling_1.m_x_Rd() # getting the bending resistance in x-direction
ceiling_1.m_y_Rd() # getting the bending resistance in y-direction
```

### Analysing "ceiling_1"

```python
# shows some infos
ceiling_1.print_info()
```

```python
# print the FE-Elements graphically
ceiling_1.print_Nachweise(Typ='V', save_file=False, markersize=10, bad_only=False, annotation=False, annotation_size=5)
ceiling_1.print_Nachweise(Typ='M', save_file=False, markersize=10, bad_only=True, annotation=False, annotation_size=5)
ceiling_1.print_Nachweise(Typ='V', save_file=False, markersize=10, bad_only=False, annotation=True, annotation_size=5)
```

```python
# Show each individual shear force check. These tables are as csv exportable
ceiling_1.Querkraftnachweise_erfuellt(sort_values=True, save_csv=False)
ceiling_1.Querkraftnachweise_nicht_erfuellt(sort_values=True, save_csv=False)
ceiling_1.Biegenachweise_erfuellt(sort_values=True, save_csv=False)
ceiling_1.Biegenachweise_nicht_erfuellt(sort_values=True, save_csv=False)
```

```python
# show some specifics of the ceiling
ceiling_1.statische_Hoehe_d3_mm
ceiling_1.statische_Hoehe_d4_mm
ceiling_1.statische_Hoehe_dv_mm
```
